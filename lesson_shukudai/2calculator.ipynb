```python
a = float(input("1つ目の数値を入力してください: "))
operator = input("演算子を入力してください (+, -, *, /, %, **): ")
b = float(input("2つ目の数値を入力してください: "))

if operator == "+":
    print(a + b)
elif operator == "-":
    print(a - b)
elif operator == "*":
    print(a * b)
elif operator == "%":
    print(a % b)
else:
    print(a ** b)
```
