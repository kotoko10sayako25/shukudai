```python
import random

# モンスターのクラス3
class Monster: # classの作成
    def __init__(self, name, HP, attack_power): # インスタンス化で使う属性の設定をする
        self.name   = name         # 名前
        self.hp     = HP           # HP
        self.attack = attack_power # 攻撃力

    # モンスターが攻撃するメソッドを定義
    def do_attack(self):
        attack_dice = random.randint(1,6) # 1-6が出るサイコロ
        #print(attack_dice)

        if attack_dice == 1:              #　サイコロが1なら攻撃力2倍
            damage = self.attack * 2
            damage_text = f"会心の一撃！プレイヤーに{damage}のダメージ"

        elif attack_dice == 2:           #　サイコロが2なら攻撃ミス
            damage = self.attack * 0
            damage_text = f"ミス！プレイヤーは攻撃を受けない"

        else:                            # それ以外は規定の攻撃
            damage = self.attack
            damage_text = f"プレイヤーに{damage}のダメージ"

        player.hp -= damage # HP-攻撃 = 残りのHPを求める

        if player.hp <= 0 :  # プレイヤーHPが0より少なくなったら、マイナスは0に修正して、ゲームオーバーを表示
            player.hp = 0
            end_text = f"～～ゲームオーバー～～"
            return(f"{self.name}の攻撃。{damage_text}。プレイヤーのHPは{player.hp}。\n{end_text}")
        else:
            return(f"{self.name}の攻撃。{damage_text}。プレイヤーのHPは{player.hp}。")

# プレイヤーのクラス
class Player:
     def __init__(self, name, HP, attack_power): # 属性の設定
        self.name     = name
        self.hp       = HP
        self.attack = attack_power # 攻撃力

# プレイヤーが攻撃するメソッドを定義
     def do_attack(self, target_monster):  # 攻撃するモンスターを引数で指定

         attack_dice = random.randint(1,6) # 1-6が出るサイコロ
         #print(attack_dice)

         if attack_dice == 1:              #　サイコロが1なら攻撃力2倍
              damage = self.attack * 2
              text = f"会心の一撃！{target_monster.name}に{damage}のダメージ"

         elif attack_dice == 2:           #　サイコロが2なら攻撃ミス
              damage = self.attack * 0
              text = f"ミス！{target_monster.name}は攻撃を受けない"

         else:                            # それ以外は規定の攻撃
              damage = self.attack
              text = f"{target_monster.name}に{damage}のダメージ"

         target_monster.hp -= damage      # HP-攻撃 = 残りのHP求める

         if target_monster.hp <= 0 :       # モンスターのHPが0より少なくなったら、マイナスは0に修正し、ターゲットモンスターを倒したことを表示
              target_monster.hp = 0
              knock_down_text = f"{target_monster.name}を倒した"
              return(f"\n☆☆{self.name}の攻撃☆☆{text}。{target_monster.name}のHPは{target_monster.hp}。{knock_down_text}\n")
         else:
              return(f"\n☆☆{self.name}の攻撃☆☆{text}。{target_monster.name}のHPは{target_monster.hp}。\n")


# クラスを元にインスタンス化。initで設定した属性に最初の情報を入力
kimeraA  = Monster("キメラA",  28, 8) # モンスター名、最初のHP、攻撃力
kimeraB  = Monster("キメラB",  28, 8)
suraimuA = Monster("スライムA", 8, 4)
suraimuB = Monster("スライムB", 8, 4)
dorakiA  = Monster("ドラキーA", 6, 3)
dorakiB  = Monster("ドラキーB", 6, 3)
player  = Player("プレイヤー",120,15)

# モンスタ達のリスト
monsters = [kimeraA, kimeraB, suraimuA, suraimuB, dorakiA, dorakiB]

while player.hp > 0 and monsters:           # プレイヤーのHPが0より大きい時は繰り返す
                                            # リストにモンスターがいる時は繰り返す（monstersと書くだけで誰かがいることを指す）
    # プレイヤーの攻撃
    target_monster = monsters[0]            # ターゲットにするのはリストの先頭
    print(player.do_attack(target_monster)) # プレイヤーの攻撃を実行。引数にターゲットを渡す
    if target_monster.hp <= 0 :             # 先頭のモンスターのHPが0以下になったら削除し、次のモンスターがターゲットになる
        monsters.pop(0)
        if not monsters:                    # 全てのモンスターがリストからいなくなったら終了
            print(f"！！敵をすべて倒した！！")
            break

    # モンスターの攻撃
    for monster in monsters:        # モンスターがリストの順番に攻撃する
        print(monster.do_attack())  # モンスターの攻撃
        if player.hp <= 0 :         # プレイヤーのHPが0以下になったら終わる
            break
```
