```python
#ver3
import os                            # ファイルやフォルダの操作に使う
import matplotlib.pyplot as plt      # グラフを作成・表示するためのライブラリ
import matplotlib.font_manager as fm # 日本語フォントを設定するために使う
from datetime import datetime        # 現在の日付や時刻を取得する
from google.colab import files       # Google Colabからファイルをダウンロードするため

# 日本語フォントをダウンロード＆インストール（初回のみ）
!apt-get -y install fonts-noto-cjk
# フォントファイルがちゃんと存在しているか確認する（TrueならOK）
print(os.path.exists("/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc"))

# 使用する日本語フォントのファイルの場所を指定
font_path = "/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc"
# 場所を指定したフォントをPythonで使えるように登録（「jp_font」という名前で使えるようにする）
jp_font = fm.FontProperties(fname = font_path)



# X軸（日付）とY軸（歩数）のデータ(手打ち)
days = ["7/7", "7/8", "7/9", "7/10", "7/11", "7/12", "7/13"]
steps = [3880, 3745, 3960, 3430, 421, 4601, 4488]

# グラフのサイズを指定（横10インチ×縦5インチ）
plt.figure(figsize=(10, 5))
# 折れ線グラフを作成。X軸＝days、Y軸＝steps、緑色
plt.plot(days, steps, linestyle='-', color='green')


# 最大値・最小値を定義
max_idx = steps.index(max(steps))
min_idx = steps.index(min(steps))
#最大値に青色のマーカー、凡例の名前は'最高歩数'
plt.scatter(days[max_idx], steps[max_idx], color='blue', s=100, label='最高歩数')
plt.scatter(days[min_idx], steps[min_idx], color='red', s=100, label='最小歩数')


# 歩数の平均を計算（合計歩数/stepの個数）
ave_steps = sum(steps) / len(steps)

# 平均値を水平な直線で表す。青色、破線、凡例の名前は'平均歩数'
plt.axhline(ave_steps, color='blue',linestyle='--',label='平均歩数')
# 凡例を日本語フォント表示
plt.legend(prop=jp_font)


# グラフのタイトル・軸ラベルを日本語で表示できるよう日本語フォントを指定
plt.title("1週間の歩数", fontproperties=jp_font)
plt.xlabel("日付",fontproperties=jp_font)
plt.ylabel("歩数",fontproperties=jp_font)

# 補助の格子線（グリッド）の表示（Falseはなし）
plt.grid(True)


# 今日の日付を取得して、ファイル名を作成
today = datetime.now().strftime('%Y%m%d%H%M%S')
filename = f'歩数グラフ_{today}.jpg'

# グラフをjpg画像として保存（画質300dpi、余白の削除）
plt.savefig(filename,dpi=300,bbox_inches='tight')

# Google Colab 上で保存した画像を、パソコンにダウンロードする
files.download(filename)
print(f"{filename}を保存しました")

# グラフをColab上にも表示
plt.show()
```
